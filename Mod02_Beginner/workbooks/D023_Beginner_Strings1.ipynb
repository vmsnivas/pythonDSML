# Low to High - II

**Problem Description:**
You are given lowercase string (A) and you have to return after converting that in upercase form.

**Input Format:**
> First and only argument containing a lowercase string S.

**Output Format:**
> You have to return uppercase form of input string.

**Example Input:**
```
Input 1:
interviewbit

Input 2:
scaler
```

**Example Output:**
```
Output 1:
INTERVIEWBIT

Output 2:
SCALER
```

````python
class Solution:
    # @param A : string
    # @return a strings
    def solve(self, A):

        return A.upper()
````

# Vowels vs Consonants - II

**Problem Description:**
You are given lowercase string (A) and you have to tell count of vowels and consonants in it.

**Input Format:**
> First and only argument containing a lowercase string S.

**Output Format:**
> You have to return an array of two elements representing count of vowels and consonants in input string respectively.

**Example Input:**
```
Input 1:
interviewbit

Input 2:
scaler
```

**Example Output:**
```
Output 1:
[5 7]

Output 2:
[2 4]
```

````python
class Solution:
    # @param A : string
    # @return a list of integers
    def solve(self, A):
        vowels = 0
        consonants = 0

        for i in A:

            if i in ['a','e','i','o','u']:
                vowels += 1
            elif i in ['b','c','d','f','g','h','j','k','l','m','n','p','q','r','s','t','v','w','x','y','z']:
                consonants += 1

        return [vowels, consonants]
````

# Reverse the word - II

**Problem Description:**
You are given lowercase string (A) and you have to return after reversing that.

**Input Format:**
> First and only argument containing a lowercase string S.

**Output Format:**
> You have to return string which is reverse form of input string.

**Example Input:**
```
Input 1:
interviewbit

Input 2:
scaler
```

**Example Output:**
```
Output 1:
tibweivretni

Output 2:
relacs
```

````python
class Solution:
    # @param A : string
    # @return a strings
    def solve(self, A):
        
        # return A[::-1]

        newString = ''

        for i in range(len(A) -1, -1, -1):

            newString += A[i]

        return newString
````
