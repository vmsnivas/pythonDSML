# Insert that

Write a program to input N numbers array from user and insert an element Y in it at specified position X.

**Input Format:**

> First line is N which means number of elements.
Second line contains N space separated integers.
Third line is X position where Y has to be inserted.
Fourth line is Y which has to be inserted.

**Output Format:**

> N+1 space separated integers of the input array after inserting the element at required position.

```python
def main():
    # YOUR CODE GOES HERE
    # Please take input and print output to standard input/output (stdin/stdout)
    # E.g. 'input()/raw_input()' for input & 'print' for output

    N = int(input())
    A = input()
    tempList = A.split()
    myList = []

    for i in tempList:
        myList.append(int(i))

    X = int(input())
    Y = int(input())

    myList.insert(X - 1, Y)

    for i in myList:
        print(i, end=' ')

if __name__ == '__main__':
    main()
```

# Linear Search - Multiple Occurences

Given an array A and an integer B, find the number of occurrences of B in A.

**Input Format:**

> Given an integer array A and an integer B.

**Output Format:**

> Return an integer, number of occurrences of B in A.

```python
class Solution:
    # @param A : list of integers
    # @param B : integer
    # @return an integer
    def solve(self, A, B):
        count = 0
        
        for i in A:
            if i == B:
                count += 1
        
        return count
```

# Search Element

You are given an integer T (number of test cases). You are given array A and an integer B for each test case. You have to tell whether B is present in array A or not.

**Input Format:**

> First line of the input contains a single integer T.
Next, each of the test case consists of 2 lines:
First line begins with an integer |A| denoting the length of array, and then |A| integers denote the array elements.
Second line contains a single integer B

**Output Format:**

> For each test case, print on a separate line 1 if the element exists, else print 0.

```python
def main():
    T = int(input())
    
    for i in range(T):
        A = input()
        tempList = A.split()
        myList = []
        result = 0
        
        for j in range(1,len(tempList)):
            myList.append(int(tempList[j]))
            
        B = int(input())
        
        for k in myList:
            if k == B:
                result = 1
                break
            else:
                result = 0
                
        print(result)
            

if __name__ == '__main__':
    main()
```

# Product of elements

**Problem Description:**

> Write a program that returns the product of all elements present in the list.

**Input Format:**

> An integer array A as the function argument.

**Output Format:**

>Product of elements in integer format

```python
def product(lst):
    ''' input:lst-List of elements in integer format
         output:Return the product as result.'''
    
    result = 1
    # YOUR CODE GOES HERE
    for i in lst:
        result *= i
        
    return result
```

# Max and Min of an Array

**Problem Description:**

> Write a program to print maximum and minimum elements of the input array A of size N where you have to take integer N and other N elements as input from the user.

**Input Format:**

> A single line representing N followed by N integers of the array A

**Output Format:**

> A single line containing two space separated integers representing maximum and minimum elements of the input array.

```python
def main():
    # YOUR CODE GOES HERE
    # Please take input and print output to standard input/output (stdin/stdout)
    # E.g. 'input()/raw_input()' for input & 'print' for output

    A = input()
    tempList = A.split()
    N = int(tempList[0])
    min = 0
    max = 0
    
    for i in range(1, len(tempList)):
        num = int(tempList[i])
        if min == 0:
            min = num
        elif num < min:
            min = num
            
        if max == 0:
            max = num
        elif num > max:
            max = num
            
    print(max, min)

if __name__ == '__main__':
    main()
```

# Right Shift an Array

**Problem Description:**
> Write a program that reads an integer array A from input and modifies the array by shifting each element to the right by one position and by shifting the last element to the first position.
Return the modified array.

**Input Format:**
> The first line of the input is the size of the array N.
The second line of the input is N integers of the array separated by space.

**Output Format:**
> Print a single line that is the elements of the modified array separated by space.

```python
def main():
    # YOUR CODE GOES HERE
    # Please take input and print output to standard input/output (stdin/stdout)
    # E.g. 'input()/raw_input()' for input & 'print' for output

    N = int(input())
    A = input()
    tempList = A.split()
    myList = []
    
    for i in range(len(tempList)):
        if i == 0:
            myList.append(int(tempList[-1]))
        else:
            myList.append(int(tempList[i - 1]))
            
    for i in myList:
        print(i, end=' ')

if __name__ == '__main__':
    main()
```

# Shopping list

**Problem Description:**

> Write a program that will keep track of items for a shopping list. The program should keep asking for new items as input until “end” is entered and also print the full shopping list.

**Input Format:**

> Input is taken from the user for different elements until end is input.

**Output Format:**

> The list of all elements of the shopping list

```python
def shopping_list():
    shoplist=[]
    ''' input:take items as inputs (strings) from the user
        output:Print the shopping list.'''
    # YOUR CODE GOES HERE

    while True:
        A = input()
        if A != 'end':
            shoplist.append(A)
        else:
            break
    
    print(shoplist)

'''Calling the function'''

shopping_list() 
```

# Print in Reverse

**Problem Description:**

> Write a program to print the input array A of size N in reverse order where you have to take integer N and further N elements as input from user.

**Input Format:**

> A single line representing N followed by N integers of the array A

**Output Format:**

> A single line containing N space separated integers representing elements of the input array in reverse order.

```python
def main():
    # YOUR CODE GOES HERE
    # Please take input and print output to standard input/output (stdin/stdout)
    # E.g. 'input()/raw_input()' for input & 'print' for output

    A = input()
    tempList = A.split()
    N = int(tempList[0])
    
    for i in range(N, 0, -1):
        print(int(tempList[i]), end=' ')

if __name__ == '__main__':
    main()
```

# Swap elements

Choose the options to complete the function swap() to swap the first and the last elements in a list?

**Input:** n_list=[12,41,29,32,7]

```python
def  swap(n_list):
    n_list[0], n_list[-1] = n_list[-1], n_list[0]
    
    return n_list

n_list = [12,41,29,32,7]
swap(n_list)
```
